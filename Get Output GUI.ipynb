```python
from tkinter import *
import tkinter as tk
from tkinter.ttk import *
from tkinter import filedialog,messagebox, ttk
from pandastable import Table, TableModel
import pandas as pd
import numpy as np
import openpyxl
```

```python
data = pd.read_excel(r'./Output/Amount to Deploy.xlsx', engine = "openpyxl")
data.head()
```

```python
def openNewWindow():
    newWindow = Toplevel(main_window)
    newWindow.title("New Window")
    Label(newWindow, text ="This is a new window").pack(pady=10)
    label = Label(main_window, 
              text ="This is the main window")
    label.pack(pady = 10)
```

```python
def search_site():
    openNewWindow()
    while True:
        loc_code = input('\nEnter the location code:-')
        sku = int(input('Enter the SKU:- '))
        search_result = data[(data['Location Code']==loc_code)&(data['SKU']==sku)]
        if search_result.empty:
            print('\n Invalid Location Code or SKU\n')
        else:
            print('\n Site Details:- \n')
            display(search_result)
            print('\n\n')
            break
```

```python
class search_site(tk.Frame):
    def __init__(self, master):
        tk.Frame.__init__(self, master)
        tk.Label(self, text="1. Search by Location Code and SKU (Single entry).").grid(row = 0, column = 1)
        Label(self, text = "").grid(row = 1, column = 1)
        Label(self, text = "Enter the location code:- ").grid(row = 2, column = 0)
        Label(self, text = "Enter the SKU:- ").grid(row = 3, column = 0)
        
        # Text Input
        s = ttk.Style()
        s.configure("b1.TButton", borderwidth=0)
        LC_code = Entry(self, width = 35, style="b1.TButton")
        LC_code.grid(row = 2 , column = 1)
        SKU_code = Entry(self, width = 35, style="b1.TButton")
        SKU_code.grid(row = 3 , column = 1)
        
        def disp():
            label = Label(self, text = "")
            label.grid(row = 5, column = 1)
            LC = LC_code.get()
            SKU = int(SKU_code.get())
            search_result = data[(data['Location Code']==LC)&(data['SKU']==SKU)]
            if search_result.empty:
                label.config(text = 'Error: Invalid Location Code or SKU.')
            else:
                master.switch_frame(TestApp(master,df=search_result))
        
        tk.Button(self, text="Display",
                  command=lambda: disp()).grid(row=6, column = 1)
        
        tk.Button(self, text="Return to start page",
                  command=lambda: master.switch_frame(StartPage)).grid(row=7, column = 1)

```

```python
class search_grid(tk.Frame):
    def __init__(self, master):
        tk.Frame.__init__(self, master)
        tk.Label(self, text="2. Search for a grid by Supply Site Code and SKU.").grid(row = 0, column = 1)
        Label(self, text = "").grid(row = 1, column = 1)
        Label(self, text = "Enter the Supply Site code:- ").grid(row = 2, column = 0)
        Label(self, text = "Enter the SKU:- ").grid(row = 3, column = 0)
        
        # Text Input
        s = ttk.Style()
        s.configure("b1.TButton", borderwidth=0)
        LC_code = Entry(self, width = 35, style="b1.TButton")
        LC_code.grid(row = 2 , column = 1)
        SKU_code = Entry(self, width = 35, style="b1.TButton")
        SKU_code.grid(row = 3 , column = 1)
        
        def disp():
            label = Label(self, text = "")
            label.grid(row = 5, column = 1)
            LC = LC_code.get()
            SKU = int(SKU_code.get())
            search_result = data[(data['Supply Site Code']==LC)&(data['SKU']==SKU)]
            if search_result.empty:
                label.config(text = 'Error: Invalid Location Code or SKU.')
            else:
                master.switch_frame(TestApp(master,df=search_result))

        
        tk.Button(self, text="Display",
                  command=lambda: disp()).grid(row=6, column = 1)
        
        tk.Button(self, text="Return to start page",
                  command=lambda: master.switch_frame(StartPage)).grid(row=7, column = 1)
```

```python
class TestApp(tk.Frame):
        def __init__(self, master,df):
            tk.Frame.__init__(self, master)
            self.main = self.master
            self.main.geometry('700x500+100+100')
            self.main.title('Table app')
            tk.Frame().grid_rowconfigure(0, weight=1)
            tk.Frame().grid_columnconfigure(0, weight=1)
            f = Frame(self.main)
            f.grid(row=0,column =0)
            self.table = pt = Table(f, dataframe=df,
                                    showtoolbar=True, showstatusbar=True, rows = 10, cols = 10)
            
            pt.show()
            tk.Button(self, text="Return to start page",
                  command=lambda: master.switch_frame(StartPage)).grid(row=1, column = 0)
     
```

```python
class SampleApp(tk.Tk):
    def __init__(self):
        tk.Tk.__init__(self)
        self._frame = None
        self.switch_frame(StartPage)

        
    def switch_frame(self, frame_class):
        """Destroys current frame and replaces it with a new one."""
        new_frame = frame_class(self)
        if self._frame is not None:
               self._frame.destroy()
        self._frame = new_frame
        self._frame.grid()

class StartPage(tk.Frame):
    def __init__(self, master):
        tk.Frame.__init__(self, master)
        self.master.title('Stock Distribution')
        Label(self, text = "Options to Search for Amount to Deploy").grid(row = 0, column = 1)
        Label(self, text = "1. Search by Location Code and SKU (Single entry).").grid(row = 1, column = 1)
        Label(self, text = "2. Search for a grid by Supply Site Code and SKU.").grid(row = 2, column = 1)
        Label(self, text = "3. Exit.").grid(row = 3, column = 1)
        Label(self, text = "Enter Option").grid(row = 5, column = 0)
        
        # Text Input 
        s = ttk.Style()
        s.configure("b1.TButton", borderwidth=0)
        e = Entry(self, width = 35, style="b1.TButton")
        e.grid(row = 5 , column = 1)
        def on_click():
            choice = e.get()
            label = Label(self, text = "")
            label.grid(row = 7, column = 1)
            if choice=='1':
                master.switch_frame(search_site)
            elif choice=='2':
                master.switch_frame(search_grid)
            elif choice =='3':
                self.destroy()
            else:
                label.config(text = 'Error: Invalid Choice.')
                

        # Buttons
        Button(self, text = "Enter", command = on_click).grid(row=6, column = 1)



if __name__ == "__main__":
    app = SampleApp()
    app.mainloop() 
```
